```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#This suppresses warning messages in Python — it tells the interpreter to ignore all warnings that would normally be printed.
df=pd.read_csv('sonar_dataset.csv',header=None)
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df[60].value_counts()
```

```python
x=df.drop(60,axis='columns')
y=df[60]
x.head()
```

```python
y=pd.get_dummies(y,drop_first=True)
y=y.astype(int)
y.sample(5)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
import tensorflow as tf
from tensorflow import keras
model=keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.fit(x_train,y_train,epochs=100,batch_size=8)
```

rn we see accuracy is 1 but when we evaluate its 73 percent only so thats how we overfit the model and then dropout regularization is needed.

```python
model.evaluate(x_test,y_test)
```

```python
y_pred=model.predict(x_test).reshape(-1)
print(y_pred[:10])
y_pred=np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,y_pred))
```

now we will dropout

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
import tensorflow as tf
from tensorflow import keras
model=keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation='sigmoid')
])
#everytime u run it u will get diff results bcz dropout layer is random.the neurons dropped are always randome so run it 10 times and 10 times u will get diff results.
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
#now accuracy aint 100% its 91% but when we evaluate accuracy would increase
model.evaluate(x_test,y_test)
```

```python
y_pred=model.predict(x_test).reshape(-1)
print(y_pred[:10])
y_pred=np.round(y_pred)
print(y_pred[:10])
print(classification_report(y_test,y_pred))
```
